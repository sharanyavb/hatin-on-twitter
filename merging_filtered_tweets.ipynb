```python
import pandas as pd
import csv
```

```python
sharanya = "Filtered_Tweets/sharanya.csv"
ronessa = "Filtered_Tweets/ronessa.csv"
mariah = "Filtered_Tweets/mariah.csv"
```

```python
num_fave = []
num_foll = []
num_frnd = []
num_hash = []
num_men = []
acc_create = []
in_reply_to = []
loc = []
name = []
acc_tweets = []
phrase = []
race = []
screen_name = []
time = []
tweet = []
user_id = []
classification = []
compound = []
racist = []
state_abb = []
valid_city_data = []
```

```python
with open(sharanya, encoding='utf-8', errors='ignore') as fdata:
    info = csv.DictReader(fdata)
    
    for row in info:
        num_fave.append(row['# favorite of tweet'])
        num_foll.append(row['# followers'])
        num_frnd.append(row['# following'])
        num_hash.append(row['# of hastags'])
        num_men.append(row['# of mentions'])
        acc_create.append(row['account creation date'])
        in_reply_to.append(row['in reply to'])
        loc.append(row['location'])
        name.append(row['name'])
        acc_tweets.append(row['number of account tweets'])
        phrase.append(row['phrase'])
        race.append(row['race'])
        screen_name.append(row['screen name'])
        time.append(row['time'])
        tweet.append(row['tweet'])
        user_id.append(row['user id'])
        classification.append(row['classification'])
        compound.append(row['compound score'])
        racist.append(row['Racist (Y/N)'])
        state_abb.append(row['State Abbreviation'])
        valid_city_data.append(row['Valid City Data (Y/N)'])
```

```python
with open(ronessa, encoding='utf-8', errors='ignore') as fdata:
    info = csv.DictReader(fdata)
    
    for row in info:
        num_fave.append(row['# favorite of tweet'])
        num_foll.append(row['# followers'])
        num_frnd.append(row['# following'])
        num_hash.append(row['# of hastags'])
        num_men.append(row['# of mentions'])
        acc_create.append(row['account creation date'])
        in_reply_to.append(row['in reply to'])
        loc.append(row['location'])
        name.append(row['name'])
        acc_tweets.append(row['number of account tweets'])
        phrase.append(row['phrase'])
        race.append(row['race'])
        screen_name.append(row['screen name'])
        time.append(row['time'])
        tweet.append(row['tweet'])
        user_id.append(row['user id'])
        classification.append(row['classification'])
        compound.append(row['compound score'])
        racist.append(row['Racist (Y/N)'])
        state_abb.append(row['State Abbreviation'])
        valid_city_data.append(row['Valid City Data (Y/N)'])
```

```python
with open(mariah, encoding='utf-8', errors='ignore') as fdata:
    info = csv.DictReader(fdata)
    
    for row in info:
        num_fave.append(row['# favorite of tweet'])
        num_foll.append(row['# followers'])
        num_frnd.append(row['# following'])
        num_hash.append(row['# of hastags'])
        num_men.append(row['# of mentions'])
        acc_create.append(row['account creation date'])
        in_reply_to.append(row['in reply to'])
        loc.append(row['location'])
        name.append(row['name'])
        acc_tweets.append(row['number of account tweets'])
        phrase.append(row['phrase'])
        race.append(row['race'])
        screen_name.append(row['screen name'])
        time.append(row['time'])
        tweet.append(row['tweet'])
        user_id.append(row['user id'])
        classification.append(row['classification'])
        compound.append(row['compound score'])
        racist.append(row['Racist (Y/N)'])
        state_abb.append(row['State Abbreviation'])
        valid_city_data.append(row['Valid City Data (Y/N)'])
```

```python
total_info = {
    '# favorite of tweet': num_fave,
    '# followers': num_foll,
    '# following': num_frnd,
    '# of hastags': num_hash,
    '# of mentions': num_men,
    'account creation date': acc_create,
    'in reply to': in_reply_to,
    'location': loc,
    'name': name,
    'number of account tweets': acc_tweets,
    'phrase': phrase,
    'race': race,
    'screen name': screen_name,
    'time': time,
    'tweet': tweet,
    'user id': user_id,
    'classification': classification,
    'compound score': compound,
    'Racist (Y/N)': racist,
    'State Abbreviation': state_abb,
    'Valid City Data (Y/N)': valid_city_data
}
```

```python
column_names = ['# favorite of tweet','# followers','# following','# of hastags','# of mentions','account creation date',
           'in reply to','location','name','number of account tweets','phrase','race','screen name','time','tweet',
           'user id','classification','compound score','Racist (Y/N)','State Abbreviation','Valid City Data (Y/N)']

info_df = pd.DataFrame(total_info, columns=column_names)
```
